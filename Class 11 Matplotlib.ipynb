```python
#Adding elements to line plots
t=np.arange(0.0,2.0,0.001)
```

```python
t
```

```python
s=1+np.cos(2*np.pi*t)
plt.plot(t,s)

plt.xlabel('time (sec)--->')
plt.ylabel('voltage in volt--->')
plt.title('Electrical values---->')
plt.grid(True)
plt.savefig('edunock.png')
plt.show()
```

```python
y=[2,6,8,7,1,5]
x=range(len(y))
width=1/1.5
plt.bar(x,y,width,color='yellow')
plt.xlabel('time (sec)--->')
plt.ylabel('voltage in volt--->')
plt.title('Electrical values---->')
plt.grid(True)
plt.show()
```

```python
#scatter graph

N=50
x=np.random.rand(N)
y=np.random.rand(N)

colors=np.random.rand(N)
area=np.pi*(15*np.random.rand(N))**2
plt.scatter(x,y,s=area,c=colors,alpha=0.8)
plt.show()
```

```python
#Pie chart 
lab='Animals','Humans','Aliens','Trees'

sizes=[15,30,45,10]   #Total=100
exp=(0,0,0.1,0)
fig1,ax1=plt.subplots()
ax1.pie(sizes,explode=exp,labels=lab,autopct='%1.2f%%',shadow=True,startangle=210)
ax1.axis('equal')
plt.show()
```
